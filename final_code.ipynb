### Linear Regression vs Logistic Regression

Linear regression has continuous values wheras logistic regression has categorical values.
Eg: You use linear regression to predict stock prices, the answer could be any value between 0 to infinity.
On the other hand, we use logistic regression to predict categorical data, for example what is the category of a particular object, Eg: What is the gender of a person (Which category). If the email is spam or not (Yes or no). Here the answer are limited it could be either yes or no. Gender could be either Male, Female, Non Binary, etc.

```python
# Importing data set
from sklearn import datasets
import numpy as np
from sklearn.linear_model import LogisticRegression

# Iris dataset predicts the flower from their height and width

iris = datasets.load_iris()

# Prints keys inside the dataset (Tells you what is inside the dataset)
print(list(iris.keys()))
```

```python
print(iris['data']) # Print what data section has inside the iris dataset. You can replace data with any key (Jo last code mai print kiya tha)
```

```python
# Print Description
print(iris['DESCR'])
```

```python
# Printing shape of the 'data' section
# Shape means dimensions of the array it could be 1 Dimensional, 2D, 3D whatever

print(iris['data'].shape)

# We get answer 150, 4 this means it has 150 rows and each row has 4 columns
# So it is a 1 X 2 matrix or a 2D array
```

```python
# Here say that from every row we only want the 3rd column
X = iris["data"][:, 3:]


# Creating a label, this is an array where if the target value = 2, then it will show true
# True because the flower is iris virginica, if we wanted to see if the flower is iris Setosa then we would have done == 0
# Check DESCR for more info

# Our computer does not understand true and false, only we stupid humans understand this.
# So we will have to convert True to 1 and False to 0 so the computer can understand
# .astype(int) does that for us
y = (iris["target"] == 2).astype(int)

print(X)
print(y)
```

```python
# Training a logistic regression calssifier

# This function performs logistic regression
classifier = LogisticRegression()
# x and y we got in the last code snippet
# This tries to fit the data we have given from the dataset and tries to predict if the flower is iris vriginica or not
# Basically here we are training the model
classifier.fit(X, y)

# Here is an example where we try to predict if a flower with petal width 1.6 is iris virginica or not
example = classifier.predict(([[1.6]]))
example2 = classifier.predict(([[2.6]]))
print(example)
print(example2)

# If we get answer as 0 this means False or the flower is not virginica
# If we get answer as 1 this means true the flower is virginica....
```
